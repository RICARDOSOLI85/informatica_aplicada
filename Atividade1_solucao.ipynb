10. Calcular a fórmula de Bhaskara para os numeros a, b e c

```julia
function f(a,b,c)
    delta = b^2 - 4*a*c 
    x1 = (-b + sqrt(delta)) / (2*a)
    x2 = (-b - sqrt(delta)) / (2*a)
    println("x1 =  $x1 e x2 = $x2")    
end
f(-1,2,10)
```
