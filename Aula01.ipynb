### 1.4 Tipos de Dados Básicos
##### Números Inteiros (Integers): Representam números inteiros sem parte fracionária. Em Julia, existem diferentes tipos de inteiros, dependendo do tamanho em bits: `Int8`, `Int16`, `Int32`, `Int64`, `Int128` (sinais) e `UInt8`, `UInt16`, `UInt32`, `UInt64`, `UInt128` (sem sinal).

```julia
x = 10  # verifica o tipo de variável 
typeof(x) 
```

```julia
y = Int32(5) # define explicitamente o tipo 
```

#### 1.4.1 Numéros de ponto flutuante (Floating-Point Numbers)
##### Representam números com parte fracionária. Os tipos mais comuns são `Float32` (precisão simples) e `Float64` (precisão dupla). `Float64` é o tipo padrão para números de ponto flutuante em Julia.

```julia
z = 3.14159 
typeof(z)
```

#### 1.4.2 Booleanos (Booleans)

```julia
a = true 
```

```julia
b = false 
```

```julia
typeof(a), typeof(b)
```

### 1.5 Operadores Lógicos 
##### && (e lógico): Retorna `true` se ambos os operandos forem `true`, caso contrário retorna `false`. || (ou lógico): Retorna `true` se pelo menos um dos operandos for `true`, caso contrário retorna `false`. ! (negação lógica): Inverte o valor booleano do operando.

```julia
a  = true 
```

```julia
b = false 
```

```julia
a && b # E lógico 
```

```julia
a || b  # Ou lógico 
```

```julia
!a # Negação 
```
