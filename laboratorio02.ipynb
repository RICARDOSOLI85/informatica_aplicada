## 2. Projeto 1: Conversão de Escalas Termométricas

**Objetivo:** Desenvolver um programa que converta uma temperatura de graus Fahrenheit (°F) para Kelvin (K).

**Fórmula:**
$$T_K = (T_F - 32) \times \frac{5}{9} + 273.15$$

**Dados para teste:**
* Ponto de ebulição da água: $212^\circ F \rightarrow 373.15 \, K$
* Sublimação de gelo seco: $-110^\circ F \rightarrow 194.26 \, K$

**Nosso programa precisa efetuar os seguintes passos:**

* 1. Pedir para o usuário digitar a temperatura de entrada em °F.
* 2. Ler a temperatura de entrada.
* 3. Calcular a temperatura em kelvin a partir da Equação (2.5).
* 4. Escrever o resultado e parar

## 3. Projeto 2: Transferência Máxima de Potência

**Problema:** Uma fonte de tensão $V = 120V$ com resistência interna $R_s = 50\Omega$ fornece energia a uma carga $R_L$. Precisamos encontrar o valor de $R_L$ que resulta na potência máxima.

**Equações:**
1. Corrente ($I$): $$I = \frac{V}{R_s + R_L}$$
2. Potência na carga ($P_L$): $$P_L = I^2 \cdot R_L$$

**Tarefa:**
1. Crie um arranjo para $R_L$ variando de $1$ a $100\Omega$ com passo de $1\Omega$.
2. Calcule os vetores de corrente e potência.
3. Identifique o valor de $R_L$ que maximiza $P_L$.
4. Plote o gráfico de $P_L$ em função de $R_L$.
